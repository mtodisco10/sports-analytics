```python
import pandas as pd
from joypy import joyplot
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
scores = pd.read_csv('data/historical_nfl_scores.csv')
scores.columns = [col.lower() for col in scores.columns]
scores['pts_diff'] = scores['home_pts'] - scores['visitor_pts']
scores['year'] = pd.to_datetime(scores.date).dt.year
```

```python
joyplot(scores.query("(pts_diff > -40) & (pts_diff < 40)"),
        by='year',
        column='pts_diff',
        colormap = cm.autumn,
        title = 'Home Team Point Differential',
        range_style = 'own',
        figsize = (10,6),
        fade = True);
```
